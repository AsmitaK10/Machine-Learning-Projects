### Outlier Detection

```python
plt.figure(figsize=(16,9))
sns.boxplot(data=x)
```

```python
x.max()
```

```python
plt.figure(figsize=(16,9))
sns.boxplot(data= x.drop(['AREA', 'CONVEX_AREA'], axis=1))
```

```python
from sklearn.ensemble import IsolationForest
```

```python
iso = IsolationForest(random_state=0, contamination=0.05)
```

```python
clean = iso.fit_predict(x,y)
```

```python
clean
```

```python
x_clean = x[clean==1]
x_clean.shape
```

```python
y_clean = y[clean==1]
y_clean.shape
```

### Resampling the data

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ros = RandomOverSampler(random_state=0)
```

```python
x_res, y_res= ros.fit_resample(x_clean, y_clean)
```

```python
x_res.shape
```

```python
sns.countplot(x=y_res)
```

```python
y_res.value_counts()
```

### Feature Selection

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
skf = SelectKBest(k=10, score_func=chi2)
```

```python
skf.fit(x,y)
```

```python
skf.get_support()
```

```python
x = x_res.iloc[:, skf.get_support()]
```

```python
x
```

### Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
pd.DataFrame(x_scaled).describe()
```

### Cross validation

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x_scaled, y_res, random_state=0, stratify=y_res)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
import warnings
warnings.filterwarnings('ignore')
```

### Decision Tree Classifier

#### Hyperparameter tuning

```python
params = {
    'criterion' : ['gini', 'entropy', 'log_loss'],
    'splitter' : ['best', 'random'],
    'random_state' : [0,1,2,3,4]
}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
```

```python
grid = GridSearchCV(DecisionTreeClassifier(),
                   param_grid=params,cv=5)
```

```python
grid.fit(x_train, y_train)
```

```python
grid.best_estimator_
```

#### Build the model

```python
dt = grid.best_estimator_
```

```python
y_pred_dt = dt.predict(x_test)
```

#### Performance Analysis

```python
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_dt)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
print(classification_report(y_test, y_pred_dt))
```

```python
print('Accuracy score:', accuracy_score(y_test, y_pred_dt)*100)
```

#### Plot the tree

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(20,20))
plot_tree(dt, fontsize=8, filled=True, rounded=True,
         feature_names=x.columns, class_names=['K','S']);
```

### Project GUI

```python
from tkinter import *
import tkinter.messagebox
import numpy as np
import matplotlib.pyplot as plt
```

```python
top = Tk()
top.title('Pistachio Categorization System')
top.geometry("1000x800")
top.resizable(0,0)

#Deserialization
classifier = joblib.load('classifier.model')
scaler = joblib.load('scaler.model')

def check():
    global a,b,c,d,e,f,g,h,i,j
    new = np.array([[a.get(), b.get(), c.get(), d.get(), e.get(), 
                     f.get(), g.get(), h.get(), i.get(), j.get()]])
    y_pred = classifier.predict(scaler.transform(new))
    if y_pred[0] == 'Kirmizi_Pistachio':
        tkinter.messagebox.showinfo('Result','Kirmizi Pistachio')
    elif y_pred[0] == 'Siit_Pistachio':
        tkinter.messagebox.showinfo('Result','Siit Pistachio')
    

a = DoubleVar()
b = DoubleVar()
c = DoubleVar()
d = DoubleVar()
e = DoubleVar()
f = DoubleVar()
g = DoubleVar()
h = DoubleVar()
i = DoubleVar()
j = DoubleVar()

Label(top, text='Pistachio Quality Detection System',
     font=('Ubuntu',25)).place(x=120,y=30)
Label(top, text='Area',
     font=('courier',16)).place(x=20,y=100)
Entry(top, textvariable=a, font=('courier',16)).place(x=280,y=100)
Label(top, text='Perimeter',
     font=('courier',16)).place(x=20,y=150)
Entry(top, textvariable=b, font=('courier',16)).place(x=280,y=150)
Label(top, text='Major Axis',
     font=('courier',16)).place(x=20,y=200)
Entry(top, textvariable=c, font=('courier',16)).place(x=280,y=200)
Label(top, text='Minor Axis',
     font=('courier',16)).place(x=20,y=250)
Entry(top, textvariable=d, font=('courier',16)).place(x=280,y=250)
Label(top, text='EQDIASQ',
     font=('courier',16)).place(x=20,y=300)
Entry(top, textvariable=e, font=('courier',16)).place(x=280,y=300)
Label(top, text='Convex Area',
     font=('courier',16)).place(x=20,y=350)
Entry(top, textvariable=f, font=('courier',16)).place(x=280,y=350)
Label(top, text='Aspect Ratio',
     font=('courier',16)).place(x=20,y=400)
Entry(top, textvariable=g, font=('courier',16)).place(x=280,y=400)
Label(top, text='Roundness',
     font=('courier',16)).place(x=20,y=450)
Entry(top, textvariable=h, font=('courier',16)).place(x=280,y=450)
Label(top, text='Compactness',
     font=('courier',16)).place(x=20,y=500)
Entry(top, textvariable=i, font=('courier',16)).place(x=280,y=500)
Label(top, text='Shape Factor',
     font=('courier',16)).place(x=20,y=550)
Entry(top, textvariable=j, font=('courier',16)).place(x=280,y=550)

Button(top, text = '--Check--',
      font=('courier',20,'bold'), command=check).place(x=700, y=300)

top.mainloop()
```
