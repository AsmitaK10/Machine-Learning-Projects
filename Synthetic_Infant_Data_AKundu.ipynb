### Feature Selection

```python
from sklearn.feature_selection import SelectKBest, chi2
skf = SelectKBest(k=18, score_func=chi2)
skf.fit(x,y)
```

```python
skf.get_support()
```

```python
x = x.iloc[:, skf.get_support()]
x
```

### Outlier Detection

```python
plt.figure(figsize=(16,9))
sns.boxplot(data=x)
```

```python
from sklearn.ensemble import IsolationForest
```

```python
iso = IsolationForest(random_state=0, contamination=0.05)
```

```python
clean = iso.fit_predict(x,y)
```

```python
clean
```

```python
x_clean = x[clean==1]
x_clean.shape
```

```python
y_clean = y[clean==1]
y_clean.shape
```

### Resampling the data

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ros = RandomOverSampler(random_state=0)
```

```python
x_res, y_res= ros.fit_resample(x_clean, y_clean)
```

```python
x_res.shape
```

```python
sns.countplot(x=y_res)
```

```python
y_res.value_counts()
```

### Cross Validation

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x_res, y_res, random_state=0, stratify=y_res)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
import warnings
warnings.filterwarnings('ignore')
```

### Decision Tree Classifier

#### Hyperparameter tuning

```python
params = {
    'criterion' : ['gini', 'entropy', 'log_loss'],
    'splitter' : ['best', 'random'],
    'random_state' : [0,1,2,3,4]
}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
```

```python
grid = GridSearchCV(DecisionTreeClassifier(),
                   param_grid=params,cv=5)
```

```python
grid.fit(x_train, y_train)
```

```python
grid.best_estimator_
```

#### Build the model

```python
dt = grid.best_estimator_
```

```python
y_pred_dt = dt.predict(x_test)
```

#### Performance Analysis

```python
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_dt)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
print(classification_report(y_test, y_pred_dt))
```

```python
print('Accuracy score:', accuracy_score(y_test, y_pred_dt)*100)
```

#### Plot the tree

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(20,20))
plot_tree(dt, fontsize=8, filled=True, rounded=True,
         feature_names=x.columns, class_names=['N','S']);
```
